# Getting net-worth

```python
url = f"https://beinsure.com/world-richest-people/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

header = soup.find_all('h2', class_='wp-block-heading')[1]
if header:
    header_text = header.get_text(strip=True)
    if f"TOP 50 World’s Richest People {current_year}" in header_text:
        net_table = soup.find('table')
        if net_table:
            rows = net_table.find_all('tr')[1:]
            data = []
            for row in rows:
                person_list = [td.get_text(strip=True) for td in row.find_all('td')]
                data.append(person_list)
            beinsure_df = pd.DataFrame(data, columns=['rank', 'name', 'net_worth_in_billions', 'company'])
else:
    print('Data Not Available')
```

```python
beinsure_df['net_worth_in_billions'] = beinsure_df['net_worth_in_billions'].str.replace('[$B]', '', regex=True).astype(float).astype(int)
```

```python
# Update the name in beinsure_df
beinsure_df.loc[beinsure_df['name'] == 'John Mars', 'name'] = 'John Franklyn Mars'
beinsure_df.loc[beinsure_df['name'] == 'Ken Griffin', 'name'] = 'Kenneth C. Griffin'
beinsure_df.loc[beinsure_df['name'] == 'Daniel Gilbert', 'name'] = 'Dan Gilbert'
beinsure_df.loc[beinsure_df['name'] == 'David Thomson', 'name'] = 'David Thomson, 3rd Baron Thomson of Fleet'
beinsure_df.loc[beinsure_df['name'] == 'MacKenzie Scott', 'net_worth_in_billions'] = 33
```

```python
beinsure_df['name'] = beinsure_df['name'].apply(lambda x: x.replace(' ', '_'))
```

```python

beinsure_df['age'] = None
# Function to get birthplace, nationality, and citizenship
def get_info(name):
    try:
        url = f'https://en.wikipedia.org/wiki/{name}'
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        table = soup.find('table', class_='infobox biography vcard') or soup.find('table', class_='infobox vcard')
        
        born = None
        nationality = None
        citizenship = None

        if table:
            rows = table.find_all('tr') 
            for row in rows:
                header = row.find('th')
                if header:
                    header_text = header.get_text(strip=True)
                    # header_text_list.append(header_text)
                    if 'Died' in header_text:
                        age_died = row.find('td')
                        if age_died:
                            text = age_died.get_text(strip=True)
                            age_match = re.search(r'\(aged\s*([^\)]+)\)', text)
                            
                            if age_match:
                                age = age_match.group(1) # this gets the full match

                    elif 'Born' in header_text:

                        birthplace_td = row.find('td')
                        birthplace_div = row.find('div')
                        age_span = row.find('span', class_ = 'noprint ForceAgeToShow')
                        age_td = row.find('td') 

                        if birthplace_td:
                            # Extract the text and get the last part, which should be the country
                            born_text = birthplace_td.get_text().split(',')[-1]
                            born = re.sub(r'\[.*?\]', '', born_text).strip()

                        elif birthplace_div:
                            born_text = birthplace_div.get_text().split(',')[-1]
                            born = re.sub(r'\[.*?\]', '', born_text).strip()

                        if age_span:
                            age_text = age_span.get_text(strip=True)
                            age = re.search(r'\d+', age_text).group()

                        elif age_td:
                            age_text = age_td.get_text(strip=True)
                            # print(age_text)
                            age_match = re.search(r'\(age\s*(\d+)(?:–\d+)?\)', age_text)
                            if age_match:
                                age = age_match.group(1)
                            
                    if 'Nationality' in header_text:
                        nationality = row.find('td').get_text()
                    if 'Citizenship' in header_text:
                        citizenship_td = row.find('td')
                        if citizenship_td:
                            # Check if there are any list items (<li>) within the 'td' element
                            citizenship_list = citizenship_td.find_all('li')
                            if citizenship_list:
                                # Extract text from list items and clean it
                                citizenship_text = ', '.join([li.get_text() for li in citizenship_list]).split(',')[-1]
                                citizenship = re.sub(r'\s*\([^)]*\)', '', citizenship_text)  # Remove text in parentheses
                            else:
                                # If there are no list items, just get the text directly from 'td'
                                citizenship_text = citizenship_td.get_text()
                                citizenship = re.sub(r'\[.*?\]', '', citizenship_text).strip()
                                
        # Determine the country based on priority
        # Set country in order of priority: nationality > citizenship > birthplace
                if nationality:
                    country = nationality
                elif citizenship:
                    country = citizenship
                else:
                    country = born

        # Use regex to find "United States" if it's part of a longer string
        # if country:
        #     match = re.search(r'United States', country)
        #     if match:
        #         country = match.group()
        return country, age 
    
    except Exception as e:
        return str(e)

# Fetch information in parallel
with concurrent.futures.ThreadPoolExecutor() as executor:
    info = list(executor.map(get_info, beinsure_df['name']))

# Unpack the list of tuples into two separate columns
beinsure_df['country'], beinsure_df['age'] = zip(*info)
# grouped_df['country'] = info

# Strip any extra spaces and replace 'U.S.' or 'American' with 'United States'
beinsure_df['country'] = beinsure_df['country'].str.strip().apply(lambda x: 'United States' if x in ['U.S.', 'US', 'American'] else x)
beinsure_df['country'] = beinsure_df['country'].str.strip().apply(lambda x: 'Italian' if x in ['Italy'] else x)
beinsure_df['country'] = beinsure_df['country'].str.strip().apply(lambda x: 'Indian' if x in ['India'] else x)
```
